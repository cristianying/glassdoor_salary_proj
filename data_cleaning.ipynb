```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('glassdoor_jobs.csv')
```

```python
#salary parsing
#company name text only
#remove job_type,industry and size
#age of company
df.head()
```

```python
df = df[df['base_salary'] != '-1']
#labmda function is just a function but not defined to be used again
salary = df['base_salary'].apply(lambda x: x.split('(')[0])
minus_Kd = salary.apply(lambda x: x.replace('HK', '').replace('K', '').replace('$',''))

df['range_salary'] = df['base_salary'].apply(lambda x: 1 if '-' in x else 0)

df['min_salary'] = minus_Kd.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = minus_Kd.apply(lambda x: int(x) if '-' not in x else int(x.split('-')[1]))
df['avg_salary'] = (df.min_salary + df.max_salary)/2

#comapny name text only
df['company_txt'] = df.apply(lambda x: x['company_name'] if x['rating'] < 0 else x['company_name'][:-4], axis = 1)

#state field
df['company_size'] = df['company_size'].apply(lambda x: x.replace('Employees', '') if 'Employees' in x else x)
df['industry'] = df['industry'].apply(lambda x: x.replace('Industry:', '') if 'Industry:' in x else x)
df['job_type'] = df['job_type'].apply(lambda x: x.replace('Job Type:', '') if 'Job Type:' in x else x)
df

df['company_age'] = df.founded.apply(lambda x: int(datetime.datetime.now().year) - x )
df.to_csv('salary_data_cleaned.csv', index = False)

pd.read_csv('salary_data_cleaned.csv')
```
